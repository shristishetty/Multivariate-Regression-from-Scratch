```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

class MultivariateRegression:
    def __init__(self):
        self.theta = None
        self.mean = None
        self.std = None

    def normalize(self, X):
        X = X.astype(float)
        X_normalized = (X - self.mean) / self.std
        return X_normalized

    def hypothesis(self, X):
        return np.dot(X, self.theta)

    def cost_function(self, X, y):
        m = len(y)
        h = self.hypothesis(X)
        J = (1 / (2 * m)) * np.sum((h - y) ** 2)
        return J

    def gradient_descent(self, X, y, learning_rate, iterations):
        m, n = X.shape
        self.theta = np.zeros((n, y.shape[1]))
        cost_history = []
        

        for _ in range(iterations):
            h = self.hypothesis(X)
            gradient = (1 / m) * np.dot(X.T, (h - y))
            self.theta -= learning_rate * gradient
            cost = self.cost_function(X, y)
            cost_history.append(cost)

        return cost_history

    def train(self, X, y, learning_rate=0.01, iterations=1000):
        self.mean = np.mean(X, axis=0)
        self.std = np.std(X, axis=0)
        X_normalized = self.normalize(X)
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        y = y.values.reshape(-1, y.shape[1])  # Ensure y is a numpy array
        cost_history = self.gradient_descent(X_normalized, y, learning_rate, iterations)
        return cost_history

    def predict(self, X):
        X_normalized = self.normalize(X)
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        return self.hypothesis(X_normalized)

def estimate(model, X_test, y_test):
    # Make predictions
    predictions = model.predict(X_test)
    
    # Calculate R-squared
    y_mean = np.mean(y_test)
    ss_total = np.sum((y_test - y_mean) ** 2)
    ss_residual = np.sum((y_test - predictions) ** 2)
    r_squared = 1 - (ss_residual / ss_total)
    
    # Calculate Mean Squared Error (MSE)
    mse = np.mean((predictions - y_test) ** 2)
    
    # Calculate Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mse)
    
    # Print evaluation metrics
    np.set_printoptions(suppress=True)
    print(bold+"R-squared:\n"+reset, r_squared*100)
    print(bold+"Mean Squared Error (MSE):\n"+reset, mse)
    print(bold+"Root Mean Squared Error (RMSE):\n"+reset, rmse)
    return r_squared

if __name__ == "__main__":
    # Load training data
    train_data = pd.read_csv('MPR6.csv')
    
    X = train_data.drop(['Price_in_thousands','Horsepower'], axis=1)
    y = train_data[['Price_in_thousands','Horsepower']]
    
   
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

    # Instantiate and train the model
    model = MultivariateRegression()
    cost_history = model.train(X_train, y_train)
    np.set_printoptions(suppress=True)

    estimate(model, X_test, y_test)
    
    predictions = model.predict(X_test)
   
```

```python
################################
import numpy as np
import pandas as pd

class MultivariateRegression:
    def __init__(self):
        self.theta = None
        self.mean = None
        self.std = None
        self.min = None
        self.max = None

    def normalize(self, X):
        X = X.astype(float)
        self.mean = np.mean(X, axis=0)
        self.std = np.std(X, axis=0)
        self.min = np.min(X, axis=0)
        self.max = np.max(X, axis=0)
        X_normalized = (X - self.mean) / self.std
        return X_normalized

    def hypothesis(self, X):
        return np.dot(X, self.theta)

    def cost_function(self, X, y):
        m = len(y)
        h = self.hypothesis(X)
        J = (1 / (2 * m)) * np.sum((h - y) ** 2)
        return J

    def gradient_descent(self, X, y, learning_rate, iterations):
        m, n = X.shape
        self.theta = np.zeros((n, y.shape[1]))
        cost_history = []

        for _ in range(iterations):
            h = self.hypothesis(X)
            gradient = (1 / m) * np.dot(X.T, (h - y))
            self.theta -= learning_rate * gradient
            cost = self.cost_function(X, y)
            cost_history.append(cost)

        return cost_history

    def train(self, X, y, learning_rate=0.001, iterations=1000):
        X_normalized = self.normalize(X)
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        y = y.values.reshape(-1, y.shape[1])  # Ensure y is a column vector
        cost_history = self.gradient_descent(X_normalized, y, learning_rate, iterations)
        return cost_history

    def predict(self, X):
        X_normalized = (X - self.mean) / self.std
        X_normalized = np.c_[np.ones(X_normalized.shape[0]), X_normalized]  # Add bias term
        return self.hypothesis(X_normalized)

def estimate(model, X_test, y_test):
    # Make predictions
    predictions = model.predict(X_test)
    
    # Calculate R-squared
    y_mean = np.mean(y_test, axis=0)
    ss_total = np.sum((y_test - y_mean) ** 2)
    ss_residual = np.sum((y_test - predictions) ** 2)
    
    # Handle division by zero
    if np.mean(ss_total) == 0:  
        r_squared = 1.0  # Perfect fit
    else:
        r_squared = 1 - (ss_residual / ss_total)  
    
    # Calculate Mean Squared Error (MSE)
    mse = np.mean((predictions - y_test) ** 2)
    
    # Calculate Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mse)
    
    # Print evaluation metrics
    print(bold+"R-squared:\n"+reset, r_squared * 100)
    print(bold+"Mean Squared Error (MSE):\n"+reset, mse)
    print(bold+"Root Mean Squared Error (RMSE):\n"+reset, rmse)
    
    return r_squared

def k_fold_cross_validation(X, y, k=19, learning_rate=0.01, iterations=1000):
    m = len(y)
    fold_size = m // k
    indices = np.random.permutation(m)
    rmse_values = []

    for i in range(k):
        start, end = i * fold_size, (i + 1) * fold_size
        val_indices = indices[start:end]
        train_indices = np.concatenate((indices[:start], indices[end:]))

        X_train, X_val = X.iloc[train_indices], X.iloc[val_indices]
        y_train, y_val = y.iloc[train_indices], y.iloc[val_indices]

        model = MultivariateRegression()
        cost_history = model.train(X_train, y_train, learning_rate, iterations)

        predictions = model.predict(X_val)
        rmse = np.sqrt(np.mean((predictions - y_val) ** 2))
        rmse_values.append(rmse)
        
        estimate(model, X_val, y_val) 
        predictions = model.predict(X_val)
        print(bold+"Predictions:\n"+reset, predictions)
        print(y_val)
   
    return np.mean(rmse_values),predictions,model

if __name__ == "__main__":
    # Load training data
    train_data = pd.read_csv('MPR6.csv')  
    
    X = train_data.drop(['Horsepower','Price_in_thousands','Fuel_efficiency'], axis=1)
    y = train_data[['Price_in_thousands','Horsepower']]
    X = pd.get_dummies(X)
    
    rmse_mean, predictions,model = k_fold_cross_validation(X, y)
    print(bold+"Average RMSE:"+reset, rmse_mean)
```
